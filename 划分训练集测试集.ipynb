划分训练集测试集



```python
import os
import shutil
import random
import  pandas as pd
```

获取类别名称


```python
dataset_path = 'animals'
daraset_name = dataset_path.split('_')[0]
print("数据集",daraset_name)
```

```python
classes = os.listdir(dataset_path)
```

```python
len(classes)
```

```python
for i in classes:
    print(i)
```

创建训练集，测试集文件夹

```python
os.mkdir(os.path.join(dataset_path,'train'))
os.mkdir(os.path.join(dataset_path,'val'))
for animal in classes:
    os.mkdir(os.path.join(dataset_path,'train',animal))
    os.mkdir(os.path.join(dataset_path,'val',animal))
```

划分训练集，测试集，移动文件


```python
teat_frac = 0.2 #测试集比例
random.seed(123)
```

```python
df = pd.DataFrame()

print('{:^18}{:^18}{:^18}'.format('类别','训练集数据个数','测试集数据个数'))
for animal in classes:
    #读取该类别的所有图像文件名
    old_dir = os.path.join(dataset_path,animal)
    images_filename = os.listdir(old_dir)
    random.shuffle(images_filename)#随机打乱

    #划分训练集测试集
    testset_number = int(len(images_filename) * teat_frac)
    testset_images = images_filename[:testset_number]
    trainset_images = images_filename[testset_number:]

    for image in testset_images:
        old_img_path = os.path.join(dataset_path,animal,image)
        new_test_path = os.path.join(dataset_path,'val',animal,image)
        print(new_test_path)
        shutil.move(old_img_path,new_test_path)#移动文件
    for image in trainset_images:
        old_img_path = os.path.join(dataset_path,animal,image)
        new_train_path = os.path.join(dataset_path,'train',animal,image)
        shutil.move(old_img_path,new_train_path)#移动文件

    #删除旧文件
    assert  len(os.listdir(old_dir)) == 0
    shutil.rmtree(old_dir)

    print('{:^18}{:^18}{:^18}'.format(animal,len(trainset_images),len(testset_images)))
    df = df.append({'class':animal,'trainset':len(trainset_images),'testset':len(testset_images)},ignore_index=True)

shutil.move(dataset_path,daraset_name+'_split')

df['total'] = df['trainset'] + df['testset']
df.to_csv('数据量统计.csv',index=False)

```
