可视化文件夹的图片


```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from mpl_toolkits.axes_grid1 import  ImageGrid

import numpy as np
import math
import os
import cv2
from tqdm import tqdm
```

指定要可视化图像的文件夹

```python
folder_path = 'animals_split/train/狼'
```

```python
N = 36
```

```python
n = math.floor(np.sqrt(N))
n
```

读取图像


```python
images = []
for each_img in os.listdir(folder_path)[:N]:
    img_path = os.path.join(folder_path,each_img).replace('\\','/')
    img_bgr = cv2.imdecode(np.fromfile(img_path, dtype=np.uint8), -1)
    img_rgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
    images.append(img_rgb)
len(images)

```

```python
fig = plt.figure(figsize=(10,10))
grid = ImageGrid(fig,111,
                 nrows_ncols=(n,n),
                 axes_pad=0.02,
                 share_all=True
                 )
for ax, im in zip(grid,images):
    im = cv2.resize(im,(720,720))
    ax.imshow(im)
    ax.axis('off')

plt.tight_layout()
plt.show()



```
